# 2. Data Preperation

1. Drop columns and rows that we don't need

```python
df.columns
```

Subsetting the columns / Keeping only columns we want

```python
df = df[['coaster_name',
       #'Length', 'Speed',
    'Location', 'Status',
       #'Opening date',
       #'Type',
    'Manufacturer',
       #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date','Opened',
       #'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'A cceleration', 'Restraints', 'Name',
    'year_introduced', 'latitude', 'longitude', 'Type_Main', 'opening_date_clean',
       #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
    'speed_mph',
       #'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean'
    ]].copy()
```

```python
df.shape
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes
```

```python
#Rename columns
df = df.rename(columns = {
    'coaster_name' : 'Coaster_Name',
    'year_introduced' : 'Year_Introduced',
    'opening_date_clean' : 'Opening_Date',
    'speed_mph' : 'Speed_mph',
    'height_ft' : 'Height_ft',
    'Inversions_clean' : 'Inversions',
    'Gforce_clean' : 'Gforce'
})
```

```python
df.head()
```

```python
#Find the missing values
df.isna().sum()
```

```python
#Find the duplicates
df.loc[df.duplicated()]
```

Fortunately, no rows are duplicated in this dataset.

```python
#Finding duplicate for each column
df.loc[df.duplicated(subset=['Coaster_Name'])].head(5)
```

```python
#Looking why the 'Crystal Beach Cyclone' is duplicated
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

*different in year introduced

```python
#Not duplicates, only select the first version
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

```python
df.shape
```

# 3. Feature Understanding

Univariate Analysis

```python
#Counts
df['Year_Introduced'].value_counts()
```

```python
#Plot top 10 years coasters were introudced
ax = df['Year_Introduced'].value_counts().head(10) \
.plot(kind='bar', title='Top 10 Years Coaster Introduced')

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
#Frequency of Coaster speeds
ax = df['Speed_mph'].plot(kind='hist', bins = 20, title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed(mph)')
```

```python
#Kernel Density Plot for Speed
ax = df['Speed_mph'].plot(kind='kde', title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed(mph)')
```

# Feature relationships

```python
#Scatter Plot / Relationship between speed vs height
df.plot(kind='scatter', x='Speed_mph', y='Height_ft', title='Coaster Speed vs Height')
plt.show()
```

```python
#Scatterplot with seaborn, Coaster Speed Speed vs Height (Year_Introduced with color)
ax = sns.scatterplot(data=df, x='Speed_mph', y='Height_ft', hue='Year_Introduced')
ax.set_title('Coaster Speed vs Height')
plt.show()
```

```python
#Pairplot
sns.pairplot(data=df,
             vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'],
             hue='Type_Main')
plt.show()
```

```python
#Correleation between these columns
df_cor = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna(ignore_index=True).corr()
```

```python
#Heatmap
sns.heatmap(data=df_cor, annot=True)
```

# 5. Ask a question aboout the data

What are the location with the fastest roller coaster?

```python
df['Location'].value_counts()
```

```python
df.query('Location != "Other"').groupby('Location')['Speed_mph']\
  .agg(['mean', 'count'])\
  .query('count >= 10')\
  .sort_values('mean')
```

```python
ax = df.query('Location != "Other"').groupby('Location')['Speed_mph']\
  .agg(['mean', 'count'])\
  .query('count >= 10')\
  .sort_values('mean')['mean']\
  .plot(kind='barh', figsize=(12,5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
